```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
pd.set_option('display.max_columns', 30)
dados = load_breast_cancer()
x = pd.DataFrame(dados.data, columns = [dados.feature_names])
y = pd.Series(dados.target)
```

```python
x.head()
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score

modelo = ExtraTreesClassifier(n_estimators = 50)
skfold = StratifiedKFold(n_splits = 5)
resultado = cross_val_score(modelo,x,y,cv = skfold)
print(resultado.mean())
```

```python
import pandas as pd
pd.set_option('display.max_columns', 89)
mushroom = pd.read_csv('/content/mushroom_dataset.csv')
```

```python
mushroom.head()
```

```python
[mushroom[c].value_counts() for c in list(mushroom.columns)]
```

```python
x = mushroom.drop('mushroom', axis = 1)
y = mushroom['mushroom']
```

```python
x_encoding = pd.get_dummies(x)
```

```python
x_encoding.head()
```

```python
pd.set_option('display.max_rows', 117)
faltantes = x_encoding.isnull().sum()
faltantes_percentual = (faltantes/len(y))*100
print(faltantes_percentual)
```

```python
y = y.replace('EDIBLE', 0)
y = y.replace('POISONOUS', 1)
```

```python
y.value_counts()
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score

modelo = ExtraTreesClassifier(n_estimators = 100, n_jobs = -1)
skfold = StratifiedKFold(n_splits = 3, shuffle = True)
resultado = cross_val_score(modelo,x_encoding,y,cv = skfold, n_jobs = -1)
print(resultado.mean())
```
